# Lightkurve basics

###### This notebook shows a few basic commands to get started with using lightkurve to analyse TESS data

###### Requirements: 
###### You need to install Python 3. I use Anaconda Python https://www.anaconda.com/products/individual
###### You will also need to install the lightkurve python module: https://docs.lightkurve.org/

```python
import matplotlib.pyplot as plt
from lightkurve import search_targetpixelfile
```

##### The following command downloads a "TargetPixelFile" (tpf) from the MAST portal.
##### You just need to know the TIC ID and the sector you want to look at

```python
tpf = search_targetpixelfile('TIC 232540264', sector=22).download()
# TOI 1247
```

##### The interact function starts an automatic plot of the light curve for selected pixels.
##### You can zoom in to parts to the light curve using the plot on the left and select different pixels on the right
##### This is very useful for looking for background signals

```python
tpf.interact()
```

##### This is the basic command for plotting the light curve using the preset aperture
##### By changing the limits you can zoom in on the part of the curve you want

```python
pipeline_lc = tpf.to_lightcurve()
pipeline_lc.scatter()
plt.xlim(1910, 1911)
plt.ylim(0.998, 1.002)
```

##### The bin method can help to show faint signals

```python
binned_lc = pipeline_lc.bin(10)
binned_lc.scatter()
plt.xlim(1910, 1911)
plt.ylim(0.999, 1.001)
```

##### It is always a good idea to check the background flux
##### This is a quick way to find false-positive signals (but this example looks good)

```python
bkg_lc = tpf.get_bkg_lightcurve()
bkg_lc.scatter()
plt.xlim(1910, 1911)
plt.ylim(0.98, 1.01)
```
